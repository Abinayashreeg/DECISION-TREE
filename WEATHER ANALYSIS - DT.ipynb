## filter the values which contains more than 24.99 relative hummidity at 3pm

```python
clean_data = data.copy()
# indha relative_humidity_3pm a classify pana poro binary ...idhula panama apdiye panomna adhu per regression
clean_data['high_humidity_label'] = (clean_data['relative_humidity_3pm']>24.99) * 1 #true or false ie 0 or 1 aanu kandupidikuro

clean_data['high_humidity_label']
```

```python
y = clean_data[['high_humidity_label' ]].copy()
y
```

```python
x = clean_data.drop('high_humidity_label' , axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 )
```

```python
from sklearn.tree import DecisionTreeClassifier #sklearnla cart (classfication and regression) idhu matu dha iruku
# sklearn packagela id3 and c4.5 ila
```

```python
## fit on train set 
```

```python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=12) # idhu vandhu cart (classfication and regression)
humidity_classifier.fit(x_train , y_train)

# humidity_classifier is the classifier model name
```

```python
y_pred = humidity_classifier.predict(x_test)
y_pred
```

```python
# visuallizig first 10 predicted values by machine
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred) * 100
```
